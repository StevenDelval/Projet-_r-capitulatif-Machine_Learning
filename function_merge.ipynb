```python
import pandas as pd
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = 200
```

```python
df_1 = pd.read_csv('./data/data_cleans.csv')
df_2 = pd.read_csv('./data/data_5k_lignes.csv')
df_3 = pd.read_csv('./data/donnees_benji.csv')
```

```python
def groupby_track (df, track='id'):

    last_col = ["track_name", "track_id"]
    mean_col = ["popularity", "duration_ms", "explicit", "track_number", "danceability", "energy", "key", "loudness", "mode", "speechiness", "acousticness", "instrumentalness", "liveness", "valence", "tempo", "time_signature"	]
    sum_col = df.drop(last_col + mean_col, axis=1).columns

    dict_agg = {}

    for col in last_col:
        dict_agg[f'{col}'] = 'last'

    for col in mean_col:
        dict_agg[f'{col}'] = 'mean'

    for col in sum_col:
        dict_agg[f'{col}'] = 'sum'

    df = df.groupby('track_' + track).agg(dict_agg)
    del df['track_' + track]
    df = df.reset_index()

    return df
```

```python
def merge_dfs(df_clean, df_merge, track='id'):

    try:
        df_merge = pd.get_dummies(df_merge, columns=['genre'], sparse=False)
    except:
        print('Pas de colonne : "genre"')

    try:
        if df_merge['explicit'][0] == False or df_merge['explicit'][0] == True:
            df_merge['explicit'] = df_merge['explicit'].apply(lambda x : 0 if x == False else 1)
    except:
        print('Pas de colonne : "explicit"')

    try:
        if df_merge['explicit'][0] == False or df_merge['explicit'][0] == True:
            df_merge['explicit'] = df_merge['explicit'].apply(lambda x : 0 if x == False else 1)
    except:
        print('Pas de colonne : "explicit"')

    df = pd.concat([df_clean, df_merge])

    try:
        df.drop(df[df['track_number'] > 100].index, axis=0, inplace=True)
    except:
        print('Pas de colonne : "track_number"')

    df = df[df_clean.columns]

    df = groupby_track(df, track)

    return df
```

```python
df = merge_dfs(df_1, df_2, 'id')

df = merge_dfs(df, df_3, 'id')

df
```
