```python
import pandas as pd
import seaborn as sns
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = 200
```

```python
df = pd.read_csv('./Projet-_recapitulatif-Machine_Learning/data/data_110k_lignes.csv')

df.shape
```

```python
df.drop([df.columns[0]], axis=1, inplace=True)
df = df[df[['genre', 'track_id']].duplicated() == False]

df.shape
```

```python
df.drop(df.select_dtypes('object').drop('genre', axis=1).columns, axis=1, inplace=True)
```

```python
df['explicit'] = df['explicit'].apply(lambda x : 0 if x == False else 1)
del df['is_local']
del df['duration_ms.1']
```

```python
df.drop(df[df['track_number'] > 100].index, axis=0, inplace=True)
```

```python
df.to_csv('./Projet-_recapitulatif-Machine_Learning/data/data_cleans.csv', index=False)
```
